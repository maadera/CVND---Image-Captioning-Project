```python
%matplotlib notebook
import re
import numpy as np
import matplotlib.pyplot as plt

def read_loss(fn):
    file = open(fn, 'r') 
    lines = file.readlines() 
    p = re.compile(r'Loss: (?P<loss>[\d.]*),')

    loss = []
    for line in lines: 
        loss.append( float(p.search(line).group('loss')) )
        
    return np.arange(len(loss)), np.array(loss)   
```

```python
plt.figure()


step, loss = read_loss('training_log.txt')

plt.plot(step/41412, loss, label='SGD, lr=0.1, batch=10')

step, loss = read_loss('training_log2.txt')
plt.plot(step/41412, loss, label='ADAM, lr=0.001, batch=10')

step, loss = read_loss('training_log3.txt')
plt.plot(step/12942, loss, label='ADAM, lr=0.001, batch=32')


plt.legend()
plt.show()
```
